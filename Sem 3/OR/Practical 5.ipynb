```python
Practical 5
# Min z = 4x1 + x2
# Subjected to:
# 3x1 + 4x2 >= 20
# x1 + 5x2 >= 15
# x1, x2 >= 0
```

```python
from scipy.optimize import linprog
obj = [4, 1]
lhs_ineq = [[-3, -4], # Red constraint left side
            [-1, -5]] # BLue constraint left side
rhs_ineq = [-20, # Red constraint right side
            -15] # Blue constraint right side
bnd = [(0, float("inf")), # Bounds of x
       (0, float("inf"))] # Bounds of y
opt = linprog(c = obj, A_ub = lhs_ineq, b_ub = rhs_ineq,
              bounds = bnd,
              method = "interior-point")
opt
```
